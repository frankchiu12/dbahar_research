```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
indexing_df = pd.read_excel("Similarity Scores 6-1-2021.xlsx", sheet_name="Name_Sheet")
new_high_df = pd.read_excel("Similarity Scores 6-1-2021.xlsx", sheet_name="New_High", header=1)

high_df = pd.merge(indexing_df, new_high_df, left_on='Indexing', right_on='2(high)', how='inner')
high_df.drop(columns=['2(high)'], inplace=True)
high_df
```

```python
openn = ['curious',	'imaginative', 'creative', 'unconventional']
consc = ['competent', 'organized', 'dutiful', 'deliberate']
extra = ['sociable', 'extrovert', 'exhibitionist', 'excitement', 'outgoing']
agree = ['trust', 'straightforward', 'altruist', 'compliant', 'modest', 'sympathetic', 'empathetic']
neuro = ['anxious', 'irritable', 'stressed', 'shy', 'vulnerable', 'moody']
```

```python
high_df['openn'] = high_df[openn].sum(axis=1)
high_df['consc'] = high_df[consc].sum(axis=1)
high_df['extra'] = high_df[extra].sum(axis=1)
high_df['agree'] = high_df[agree].sum(axis=1)
high_df['neuro'] = high_df[neuro].sum(axis=1)

ocean_columns = ['openn', 'consc', 'extra', 'agree', 'neuro']
ocean_traits = {
    'openn': ['curious', 'imaginative', 'creative', 'unconventional'],
    'consc': ['competent', 'organized', 'dutiful', 'deliberate'],
    'extra': ['sociable', 'extrovert', 'exhibitionist', 'excitement', 'outgoing'],
    'agree': ['trust', 'straightforward', 'altruist', 'compliant', 'modest', 'sympathetic', 'empathetic'],
    'neuro': ['anxious', 'irritable', 'stressed', 'shy', 'vulnerable', 'moody']
}
high_df = high_df[['CEO Name'] + list(ocean_traits.keys())]
high_df.columns = ['CEO Name'] + ocean_columns
high_df
```

```python
result_df = pd.read_csv('/Users/franksi-unchiu/Desktop/personality/results.csv')

result_df['CEO'] = result_df['CEO'].str.lower()
high_df['CEO Name'] = high_df['CEO Name'].str.lower()
result_df.rename(columns={'CEO': 'CEO Name'}, inplace=True)

merged_df = pd.merge(result_df, high_df, on='CEO Name')
personality_traits = ['openn', 'consc', 'extra', 'agree', 'neuro']

for trait in personality_traits:
    trait_x = f'{trait}_x'
    trait_y = f'{trait}_y'

    plt.figure(figsize=(10, 6))
    plt.plot(merged_df['CEO Name'], merged_df[trait_x], marker='o', label=f'{trait}_x')
    plt.plot(merged_df['CEO Name'], merged_df[trait_y], marker='o', label=f'{trait}_y')
    plt.title(f'{trait.capitalize()} Comparison: {trait}_x vs. {trait}_y')
    plt.xlabel('CEO Name')
    plt.ylabel('Score')
    plt.legend()
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
for trait in personality_traits:
    trait_x = f'{trait}_x'
    trait_y = f'{trait}_y'

    plt.figure(figsize=(10, 6))
    sns.regplot(x=merged_df[trait_x], y=merged_df[trait_y], scatter_kws={'s':50}, line_kws={'color':'red'})
    plt.title(f'{trait.capitalize()} Comparison: {trait}_x vs. {trait}_y')
    plt.xlabel(f'{trait}_x')
    plt.ylabel(f'{trait}_y')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
